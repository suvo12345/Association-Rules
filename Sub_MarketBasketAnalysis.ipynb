```python
import pandas as pd
import mysql.connector
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fact_order_detail = pd.read_csv('BasketData.csv')
```

```python
fact_order_detail.head(10)
```

```python
basket = (fact_order_detail[fact_order_detail['OrderDetail_QTY'] > -1]
          .groupby(['E_OrderNo', 'ProductClass'])['OrderDetail_QTY']
          .sum().unstack().reset_index().fillna(0)
          .set_index('E_OrderNo'))
```

```python
fact_order_detail.ProductClass.unique().size
```

```python
fact_order_detail.ProductClass.value_counts()
```

```python
fact_order_detail.columns
```

```python
basket.head()
```

```python
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1

basket_sets = basket.applymap(encode_units)
```

```python
basket_sets.head()
```

```python
frequent_itemsets = apriori(basket_sets, min_support=0.001, use_colnames=True)
```

```python
frequent_itemsets.sort_values(by=['support'],ascending=[True]).head()
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules.head()
```
